```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['mathtext.fontset'] = 'cm' # Example

# Load the CSV file
# df = pd.read_csv("grid_search_results_tuneaphi.csv")
df = pd.read_csv("grid_search_results_v1.csv")
df

# Clean mae_phi column if it's in string format with brackets
if df['mae_phi'].dtype == object:
    df['mae_phi'] = df['mae_phi'].apply(lambda x: float(x.strip("[]")))

# Marginalize over ic_weight (group by physics_weight)
marg_physics = df.groupby('physics_weight')['mae_phi'].mean()
plt.figure(dpi=120)
marg_physics.plot(marker='o')
plt.xlabel(r'$\lambda_{\mathrm{physics}}$')
plt.ylabel(r'Average MAE of $\phi(t)$')
plt.title(r'MAE($\phi$) vs $\lambda_{\mathrm{physics}}$ (marginalized over $\lambda_{a}$)')
# plt.title(r'$\lambda_{\mathrm{p1a}}$')
# plt.title(r'MAE($\phi$) vs $\lambda_{a}$')
plt.grid(True)
plt.tight_layout()
plt.savefig('marginalizedMAE_aw_test.png')
plt.show()

# Marginalize over physics_weight (group by ic_weight)
marg_ic = df.groupby('ic_weight')['mae_phi'].mean()
plt.figure(dpi=120)
marg_ic.plot(marker='s', color='orange')
plt.xlabel(r'$\lambda_{\mathrm{IC}}$')
plt.ylabel(r'Average MAE of $\phi(t)$')
plt.title(r'MAE($\phi$) vs $\lambda_{\mathrm{IC}}$ (marginalized over $\lambda_{a}$)')
# plt.title(r'MAE($\phi$) vs $\lambda_{\phi}$')
plt.grid(True)
plt.tight_layout()
plt.savefig('marginalizedMAE_phiw_test.png')
plt.show()
```
